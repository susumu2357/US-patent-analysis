```python
import pandas as pd
office_action = pd.read_csv("../data/office_actions.csv")
```

```python
citation = pd.read_csv("../data/citations.csv")
```

```python
merge_cite_oa =  pd.merge(citation, office_action, on = 'ifw_number').drop('app_id_y', axis = 1).rename(columns = {'app_id_x':'app_id'})
```

```python
merge_cite_oa
```

```python
app2017 = pd.read_csv('../data/app2017_pgpub_correspondence.csv', index_col = 0)
```

```python
app2017 = app2017.drop(['pgpub_num', '1'], axis = 1).rename(columns = {'app_num':'app_id'})
```

```python
app2017
```

```python
cite_oa_app2017 = pd.merge(merge_cite_oa, app2017, on = 'app_id')
```

```python
cite_oa_app2017
```
